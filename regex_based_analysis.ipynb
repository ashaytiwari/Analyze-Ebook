# Load the book

```python
with open("miracle_in_the_andes.txt", "r", encoding="utf-8") as file:
    book = file.read()
```

# How many chapters in the book?

### With string methods

```python
print(book.count("Chapter"))
```

### With regex

```python
import re

pattern = re.compile("Chapter [0-9]+")

chapter_list = re.findall(pattern, book)
chapter_count = len(chapter_list)

print(f'Chapters List: {chapter_list}')
print(f'Total number of chapters: {chapter_count}')
```

# Which are the sentences where "love" was used?

```python
pattern = re.compile("[^.]*[^a-zA-Z]+love[^a-zA-Z]+[^.]*.")
sentences = re.findall(pattern, book)
print(sentences)
print(len(sentences))
```

# What are the most used words?

```python
pattern = re.compile("[a-zA-z]+")
findings = re.findall(pattern, book.lower())
```

```python
dictionary = {}

for word in findings:
    if word in dictionary.keys():
        dictionary[word] = dictionary[word] + 1
    else:
        dictionary[word] = 1
```

```python
dictionary_list = [(value, key) for (key, value) in dictionary.items()]
sorted(dictionary_list, reverse=True)
```
