# Load the book

```python
with open("miracle_in_the_andes.txt", "r", encoding="utf-8") as file:
    book = file.read()
```

# The most used words (non-articles)

```python
import re

pattern = re.compile("[a-zA-z]+")
findings = re.findall(pattern, book.lower())
```

```python
dictionary = {}

for word in findings:
    if word in dictionary.keys():
        dictionary[word] = dictionary[word] + 1
    else:
        dictionary[word] = 1
```

```python
dictionary_list = [(value, key) for (key, value) in dictionary.items()]
dictionary_list = sorted(dictionary_list, reverse=True)
```

```python
import nltk

from nltk.corpus import stopwords

english_stopwords = stopwords.words("english")
```

```python
filtered_list = []

for count, word in dictionary_list:
    if word not in english_stopwords:
        filtered_list.append((word, count))
```

# Sentiment Analysis: What is the most positive and negative chapter?

```python
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
analyzer = SentimentIntensityAnalyzer()
```

### An Example

```python
analyzer.polarity_scores("Hey look, How beautiful this world is? I love this nature.")
```

```python
analyzer.polarity_scores("This society is shit, how bad these people are? They don't have basic humanity. I wish to kill them all. I will burn them to ashes.")
```

### Chapter Sentiment Analysis

```python
import re
pattern = re.compile("Chapter [0-9]+")
chapters = re.split(pattern, book)
chapters = chapters[1:]
```

```python
score_list = []

for index, chapter in enumerate(chapters):
    scores = analyzer.polarity_scores(chapter)
    score_list.append((f'Chapter {index + 1}', scores))

score_list
```

```python
most_negative_chapter = ''
most_negative_polarity_score = 0

most_positive_chapter = ''
most_positive_polarity_score = 0

for (key, value) in score_list:
    
    # find most negative chapter
    if value["neg"] > value["pos"]:
        
        if value["neg"] > most_negative_polarity_score:
            most_negative_chapter = key
            most_negative_polarity_score = value["neg"]

    # find most positive chapter
    else:
        if value["pos"] > most_positive_polarity_score:
            most_positive_chapter = key
            most_positive_polarity_score = value["pos"]

print(f"Most Negative Chapter is {most_negative_chapter} with score: {most_negative_polarity_score}")
print(f"Most Positive Chapter is {most_positive_chapter} with score: {most_positive_polarity_score}")
            
```
